## Question:
### Does the test preparation course have any affect on the student's average exam score?

```python
import pandas as pd
import numpy as np
```

```python
# df = pandas Dataframe that is read in by the pandas package

df = pd.read_csv("exams.csv")
```

```python

df["average scores"] = ((df["math score"] + df["reading score"] + df["writing score"])/3).round(1)
#creates new column "average scores" which is the average of all 3 exam scores for the student
df.head(5)
# shows first 5 entries
```

```python
import plotly.express as px

fig = px.scatter(df, x=df.index, y="average scores", color="test preparation course", trendline="ols")
#gives the average score of all 3 exams for each student index value, color orange for no test prep and purple for test prep completed
fig.show()
```

```python
import plotly.express as px

fig = px.box(df, x="average scores", color="test preparation course")
fig.update_traces(quartilemethod="exclusive") # or "inclusive", or "linear" by default
fig.show()
```

## Findings:

I wanted to find out from my data whether or not the test preparation course had any effect on the
student's average exam scores. To find the average exam scores, I created a new calculated column
which added all 3 exam scores together and divided by the number of exams that were taken (3). The
student then had one of two values under test preparation course: completed or none. I imported and 
used plotly to create my two visualizations.

My first visualization shows each index value's (student) average exam score. There were 1000 students in my dataset. The visualization shows that the purple (completed test preparation course) trendline is higher than the orange for which students did not complete a test preparation course. In my second box plot visualization, we can also see that the students that completed a test preparation course (purple) had better exam results than the students who did not (orange). Therefore, it is safe to assume that the test preparation course contributes to better student exam scores based off of our 
data.
